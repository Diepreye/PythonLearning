```python
#It's hard to modify a global variable. To do this "global variable_name"
#This shouldn't be done often as it can be confusing
#Avoid modifying global scope within a function that has local scope
#A return statement can be used instead

#Number Guessing Game Objectives:

# Include an ASCII art logo.
import random
logo = """
   ___                      _____  _           _  _               _              
  / __| _  _  ___  ___ ___ |_   _|| |_   ___  | \| | _  _  _ __  | |__  ___  _ _ 
 | (_ || || |/ -_)(_-<(_-<   | |  | ' \ / -_) | .` || || || '  \ | '_ \/ -_)| '_|
  \___| \_,_|\___|/__//__/   |_|  |_||_|\___| |_|\_| \_,_||_|_|_||_.__/\___||_|  
                                      """     
print(logo)                              
actual_number = random.randint(1,100)
difficulty_dict = {'e':'easy','h':'hard'}
print("I'm thinking of a random number between 1 and 100")
print("Can you guess the number? Then have a go...")
difficulty = input("Choose a difficulty level: Type 'e' for easy or 'h' for hard\n")
def number_of_lives(): 
  if difficulty == "e":
    lives = 10
    return lives
  else:
    lives = 5
    return lives

lives = number_of_lives()
print(f"You've decoded to go {difficulty_dict[difficulty]} and you have {lives} attempts")
guess_list = [] 
continue_guessing = True
while continue_guessing == True:
  player_guess = int(input("Please input a number between 1 and 100: "))
  if player_guess > actual_number:
    print("Too high")
  elif player_guess < actual_number:
    print("Too low")
  if player_guess > actual_number or player_guess < actual_number:
    lives -= 1
    guess_list.append(player_guess)
    print(f"Here's a list of the numbers you've guessed already: {guess_list}")
    print("Guess again")
    print(f'You have {lives} attempts left')
    if lives == 0:
      print("Oops! You've lost all your lives")
      print(f'The right number is {actual_number}')
      continue_guessing = False
  else:
    print("You guessed the right number. You win!")
    continue_guessing = False
```
